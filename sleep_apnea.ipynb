```python
from hmmlearn import hmm
import numpy as np

np.random.seed(42)
n_samples = 1000
n_features = 3
X = np.random.randn(n_samples, n_features)

# Create a GaussianHMM model with 3 hidden states
model = hmm.GaussianHMM(n_components=3)

# Train the model
model.fit(X)

# Access the learned model parameters
# Transition matrix
print("Learned Transition Matrix:")
print(model.transmat_)

# Emission matrix (means)
print("Learned Emission Means:")
print(model.means_)

# Covariance matrices (variances)
print("Learned Covariance Matrices:")
print(model.covars_)
```

```python
# Predict the most likely hidden states for a sequence of observations
X_test = np.random.randn(10, n_features)  # Example test data
hidden_states = model.predict(X_test)
print("Predicted Hidden States:")
print(hidden_states)
```

```python
# Compute the log-likelihood of a sequence of observations
log_likelihood = model.score(X_test)
print("Log-Likelihood of Observations:", log_likelihood)

# Decode the most likely hidden state sequence for a sequence of observations
hidden_states = model.decode(X_test)[1]
print("Decoded Hidden States:")
print(hidden_states)

# Compute the posterior probabilities of hidden states given observations
posterior_probs = model.predict_proba(X_test)
print("Posterior Probabilities of Hidden States:")
print(posterior_probs)
```

```python
X_generated, hidden_states_generated = model.sample(10)
print("Generated Observations:")
print(X_generated)

print("Generated Hidden States:")
print(hidden_states_generated)
```

```python
import numpy as np
from hmmlearn import hmm

np.random.seed(0)
n_samples = 1000
obs_len = 10
n_states = 3
X = np.random.randn(n_samples, obs_len)

# Define the HMM model
model = hmm.GaussianHMM(n_components=n_states, n_iter=100, tol=0.01)

# Fit the model using the EM algorithm
model.fit(X)

# Print the learned parameters
print("Learned Transition Matrix:")
print(model.transmat_)
print("Learned Means of Gaussian Emission Distribution:")
print(model.means_)
print("Learned Covariance Matrices of Gaussian Emission Distribution:")
print(model.covars_)
```

```python
import numpy as np
from hmmlearn import hmm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

model = hmm.GaussianHMM(n_components=3)  
X = np.random.randn(100, 3) 
model.fit(X)

obs_seq, _ = model.sample(6)  

hidden_states = model.predict(obs_seq)

ground_truth_labels = np.array([0, 1, 2, 1, 0, 2]) 

# Calculate accuracy
accuracy = accuracy_score(ground_truth_labels, hidden_states)

# Calculate precision
precision = precision_score(ground_truth_labels, hidden_states, average='weighted')

# Calculate recall
recall = recall_score(ground_truth_labels, hidden_states, average='weighted')

# Calculate F1-score
f1_score = f1_score(ground_truth_labels, hidden_states, average='weighted')

# Print the calculated metrics
print("Accuracy: ", accuracy)
print("Precision: ", precision)
print("Recall: ", recall)
print("F1-score: ", f1_score)
```
